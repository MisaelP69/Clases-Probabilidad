## Problema de Probabilidad Condicional

Suponga que el espacio muestral es la población de adultos en una pequeña ciudad que cumplen con los requisitos para obtener un grado en la universidad. Se clasifican de acuerdo con su sexo y situación laboral. La información se presenta en la siguiente tabla:

|               | Empleado | Desempleado |
|---------------|----------|-------------|
| **Hombre**    | 460      | 40          |
| **Mujer**     | 140      | 260         |

Un individuo será seleccionado al azar para realizar un viaje a través del país para promover las ventajas de establecer industrias nuevas en la ciudad.

---

### **Pregunta:**  
¿Cuál es la probabilidad de que el individuo seleccionado sea **hombre**, sabiendo que **tiene empleo**?

---

### Resolución

Queremos calcular la **probabilidad condicional**:

$$
P(\text{Hombre} \mid \text{Empleado}) = \frac{P(\text{Hombre y Empleado})}{P(\text{Empleado})}
$$

---

### **Paso 1: Total de personas empleadas**

$$
\text{Total empleados} = 460 \text{ (hombres)} + 140 \text{ (mujeres)} = 600
$$

---

### **Paso 2: Número de hombres empleados**

$$
P(\text{Hombre y Empleado}) = \frac{460}{1000}
$$

---

### **Paso 3: Aplicar la fórmula**

$$
P(\text{Hombre} \mid \text{Empleado}) = \frac{460}{600} \approx 0.767
$$

---

### ✅ **Respuesta final**

$$
\boxed{P(\text{Hombre} \mid \text{Empleado}) = \frac{460}{600} = 0.767}
$$

**Interpretación:**  
Si se selecciona una persona empleada al azar, hay un **76.7%** de probabilidad de que sea **hombre**.

## Probabilidad de resultado específico con larvas

Se aplica un insecticida a **tres larvas** y al cabo de 24 horas se observa si están **muertas o vivas**.

---

### **Pregunta:**  
¿Cuál es la probabilidad de que haya **dos larvas muertas y una viva**?

---

### **Supuestos:**
- Cada larva muere **de manera independiente** con una probabilidad \( p \).
- Entonces, la probabilidad de que una larva esté viva es \( 1 - p \).

Queremos encontrar:

$$
P(\text{2 muertas y 1 viva})
$$

---

### **Número de combinaciones posibles:**

Hay 3 formas diferentes de que ocurra el resultado "dos muertas y una viva":  
- Muerta, muerta, viva  
- Muerta, viva, muerta  
- Viva, muerta, muerta  

Esto se puede calcular con el **coeficiente binomial**:

$$
\binom{3}{2} = 3
$$

---

### **Probabilidad de un solo resultado (por ejemplo, muerta, muerta, viva):**

$$
P = p \cdot p \cdot (1 - p) = p^2 (1 - p)
$$

---

### **Probabilidad total (tres combinaciones):**

$$
P(\text{2 muertas y 1 viva}) = 3 \cdot p^2 \cdot (1 - p)
$$

---

### ✅ **Resultado general:**

$$
\boxed{P = 3 \cdot p^2 \cdot (1 - p)}
$$

---

### 🔢 **Ejemplo numérico (si \( p = 0.7 \)):**

$$
P = 3 \cdot (0.7)^2 \cdot (1 - 0.7) = 3 \cdot 0.49 \cdot 0.3 = 0.441
$$

**Interpretación:**  
Si la probabilidad de muerte de cada larva es 0.7, entonces hay una probabilidad de **44.1%** de que exactamente **dos mueran y una viva**.

0.6 = P(E) + 0.4 - 0.1

### ✅ **Respuesta final:**

La probabilidad de que el estudiante apruebe **Estadística** es **30%**.


## Probabilidad de seleccionar dos fusibles defectuosos (sin reemplazo)

Supóngase que se tiene una **caja de 20 fusibles**, de los cuales **5 están defectuosos**.  
Se seleccionan **2 fusibles al azar sin reemplazo**.

---

### **Pregunta:**
¿Cuál es la probabilidad de que **ambos fusibles** seleccionados resulten **defectuosos**?

---

### Resolución:

Este es un caso de probabilidad compuesta **sin reemplazo**, por lo que calculamos:

$$
P(\text{defectuoso}_1 \cap \text{defectuoso}_2) = P(\text{defectuoso}_1) \cdot P(\text{defectuoso}_2 \mid \text{defectuoso}_1)
$$

---

### **Paso 1: Probabilidad de que el primer fusible sea defectuoso**

$$
P_1 = \frac{5}{20}
$$

---

### **Paso 2: Probabilidad de que el segundo fusible sea defectuoso dado que el primero lo fue**

Ahora quedan **4 defectuosos de 19 fusibles restantes**:

$$
P_2 = \frac{4}{19}
$$

---

### **Paso 3: Multiplicamos ambas probabilidades**

$$
P = \frac{5}{20} \cdot \frac{4}{19} = \frac{1}{4} \cdot \frac{4}{19} = \frac{4}{76} = \frac{1}{19} \approx 0.0526
$$

---

### ✅ **Respuesta final:**

$$
\boxed{P = \frac{1}{19} \approx 0.0526}
$$

**Interpretación:**  
La probabilidad de que ambos fusibles seleccionados al azar estén defectuosos es aproximadamente **5.26%**.


## Teorema de Bayes: Chocolatinas

Un niño tiene **3 cajas** con chocolatinas:

- **Caja 1**: 2 negras y 1 blanca  
- **Caja 2**: 2 negras y 2 blancas  
- **Caja 3**: 1 negra y 2 blancas  

Su hermana pequeña coge **una chocolatina blanca al azar**, sin saber de qué caja proviene.

---

### **Pregunta:**
¿Cuál es la probabilidad de que haya cogido la chocolatina blanca de la **primera caja**, dado que fue blanca?

---

### **Usamos el Teorema de Bayes:**

Sea:  
- \( A_1 \): la chocolatina fue tomada de la **Caja 1**  
- \( B \): se ha tomado una **chocolatina blanca**

Queremos calcular:  
$$
P(A_1 \mid B) = \frac{P(B \mid A_1) \cdot P(A_1)}{P(B)}
$$

---

### **Paso 1: Probabilidades iniciales**

- \( P(A_1) = P(A_2) = P(A_3) = \frac{1}{3} \) (las cajas son igualmente probables)
- \( P(B \mid A_1) = \frac{1}{3} \) (1 blanca de 3 total)
- \( P(B \mid A_2) = \frac{2}{4} = \frac{1}{2} \)
- \( P(B \mid A_3) = \frac{2}{3} \)

---

### **Paso 2: Probabilidad total de sacar una chocolatina blanca**

$$
P(B) = P(B \mid A_1) \cdot P(A_1) + P(B \mid A_2) \cdot P(A_2) + P(B \mid A_3) \cdot P(A_3)
$$

$$
P(B) = \frac{1}{3} \cdot \frac{1}{3} + \frac{1}{3} \cdot \frac{1}{2} + \frac{1}{3} \cdot \frac{2}{3}
$$

$$
P(B) = \frac{1}{9} + \frac{1}{6} + \frac{2}{9} = \frac{1 + 3/2 + 2}{9} = \frac{1}{9} + \frac{3}{18} + \frac{4}{18} = \frac{13}{18}
$$

---

### **Paso 3: Aplicar Teorema de Bayes**

$$
P(A_1 \mid B) = \frac{\frac{1}{3} \cdot \frac{1}{3}}{\frac{13}{18}} = \frac{1/9}{13/18} = \frac{1}{9} \cdot \frac{18}{13} = \frac{2}{13} \approx 0.154
$$

---

### ✅ **Respuesta final:**

$$
\boxed{P(\text{Caja 1} \mid \text{blanca}) = \frac{2}{13} \approx 0.154}
$$

**Interpretación:**  
Dado que se ha cogido una chocolatina blanca, la probabilidad de que haya sido de la **primera caja** es **0.154** (15.4%).


# Teoría y Solución del Problema de las Monedas Alternadas
# ======================================================

## **Enunciado del Problema**
'''
Tenemos:
- Moneda $A$: Probabilidad de cara = $s$
- Moneda $B$: Probabilidad de cara = $t$

Se lanzan alternadamente (A, B, A, B, ...).
¿Cuál es la probabilidad de que la **primera cara** sea obtenida con $A$?
'''

## **Solución Teórica**
'''
### **Ecuación Recursiva:**
Definimos $P$ como la probabilidad buscada. Puede ocurrir en dos casos:

1. **$A$ sale cara en el primer lanzamiento**:
   $$P_1 = s$$

2. **Ambas fallan ($A$ y $B$ salen cruz)** y luego se reinicia el proceso:
   - Probabilidad de fallo de $A$: $1 - s$
   - Probabilidad de fallo de $B$: $1 - t$
   - La probabilidad desde este punto sigue siendo $P$

   $$P_2 = (1 - s)(1 - t)P$$

Combinando ambos casos:
$$P = s + (1 - s)(1 - t)P$$

### **Resolución:**
$$P - (1 - s)(1 - t)P = s$$
$$P \left[1 - (1 - s)(1 - t)\right] = s$$
$$P = \frac{s}{1 - (1 - s)(1 - t)}$$

Simplificando el denominador:
$$1 - (1 - s)(1 - t) = s + t - st$$

**Resultado final:**
$$P = \boxed{\dfrac{s}{s + t - st}}$$
'''



```python

from IPython.display import display, Math  # Para mostrar ecuaciones

# Definimos probabilidades
s = 0.6
t = 0.4

# Fórmula teórica
P_teorico = s / (s + t - s * t)
display(Math(fr"P = \frac{{{s}}}{{{s} + {t} - {s} \times {t}}} = {P_teorico:.4f}"))

# Simulación (opcional)
import random
def primera_cara_A(s, t, n=100000):
    count = 0
    for _ in range(n):
        lanzamiento = 0  # 0 = A, 1 = B
        while True:
            if lanzamiento == 0:
                if random.random() < s:
                    count += 1
                    break
            else:
                if random.random() < t:
                    break
            lanzamiento = 1 - lanzamiento
    return count / n

P_sim = primera_cara_A(s, t)
display(Math(fr"\text{{Simulación}} \approx {P_sim:.4f}"))
```

# Probabilidad de no contraer enfermedades I y II
# ================================================

## **Enunciado del Problema**
'''
En una población:
- $P(\text{I}) = 10\%$ contraerá la enfermedad I.
- $P(\text{II}) = 15\%$ contraerá la enfermedad II.
- $P(\text{I} \cap \text{II}) = 3\%$ contraerá ambas.

**Objetivo:**  
Calcular la probabilidad de no contraer ninguna enfermedad.
'''

## **Solución Teórica**
'''
### **Paso 1: Definir eventos**
- $A$: Contraer enfermedad I → $P(A) = 0.10$
- $B$: Contraer enfermedad II → $P(B) = 0.15$
- Ambas → $P(A \cap B) = 0.03$

### **Paso 2: Probabilidad de al menos una enfermedad**
Por inclusión-exclusión:
$$ P(A \cup B) = P(A) + P(B) - P(A \cap B) $$
$$ P(A \cup B) = 0.10 + 0.15 - 0.03 = 0.22 \quad (22\%) $$

### **Paso 3: Probabilidad de ninguna enfermedad**
Evento complementario:
$$ P(\text{Ninguna}) = 1 - P(A \cup B) $$
$$ P(\text{Ninguna}) = 1 - 0.22 = 0.78 \quad (78\%) $$

### **Respuesta Final:**
La probabilidad de no contraer ninguna enfermedad es:
$$ \boxed{78\%} $$
'''

## **Verificación con Diagrama de Venn**


```python
from IPython.display import display, Math

# Cálculo teórico
P_A = 0.10
P_B = 0.15
P_A_and_B = 0.03

P_A_or_B = P_A + P_B - P_A_and_B
P_none = 1 - P_A_or_B

display(Math(fr"""
\text{{Probabilidad de ninguna enfermedad}} = 1 - P(A \cup B) \\
= 1 - [P(A) + P(B) - P(A \cap B)] \\
= 1 - [{P_A} + {P_B} - {P_A_and_B}] \\
= 1 - {P_A_or_B} \\
= \boxed{{{P_none*100:.0f}\%}}
"""))

# Desglose con Venn (opcional)
only_A = P_A - P_A_and_B
only_B = P_B - P_A_and_B
both = P_A_and_B
none = P_none

display(Math(fr"""
\text{{Desglose:}} \\
\text{{Solo I}} = {only_A*100:.0f}\% \\
\text{{Solo II}} = {only_B*100:.0f}\% \\
\text{{Ambas}} = {both*100:.0f}\% \\
\text{{Ninguna}} = \boxed{{{none*100:.0f}\%}}
"""))
```

# Probabilidad de al menos un bolígrafo rojo al extraer 3 sin reposición
# =====================================================================

## **Enunciado del Problema**
'''
Una caja contiene:
- $4$ bolígrafos negros ($N$),
- $3$ bolígrafos azules ($A$),
- $3$ bolígrafos rojos ($R$).

Se extraen $3$ bolígrafos **sin reposición**.  
**Calcular:** Probabilidad de que al menos uno sea rojo.
'''

## **Solución Teórica**
'''
### **Método 1: Enfoque Complementario (Recomendado)**
Calculamos la probabilidad de que **ninguno** sea rojo y restamos de 1.

1. **Total de bolígrafos no rojos:**  
   $$4 (N) + 3 (A) = 7$$

2. **Combinaciones sin rojos:**  
   $$\binom{7}{3} = \frac{7!}{3!4!} = 35$$

3. **Total de combinaciones posibles:**  
   $$\binom{10}{3} = \frac{10!}{3!7!} = 120$$

4. **Probabilidad de ningún rojo:**  
   $$P(\text{Ningún } R) = \frac{35}{120} = \frac{7}{24}$$

5. **Probabilidad de al menos 1 rojo:**  
   $$P(\text{Al menos 1 } R) = 1 - \frac{7}{24} = \boxed{\dfrac{17}{24} \approx 70.83\%}$$

### **Método 2: Enfoque Directo (Verificación)**
Sumamos las probabilidades de los casos favorables:

- **1 rojo + 2 no rojos:**  
  $$\binom{3}{1}\binom{7}{2} = 3 \times 21 = 63$$

- **2 rojos + 1 no rojo:**  
  $$\binom{3}{2}\binom{7}{1} = 3 \times 7 = 21$$

- **3 rojos:**  
  $$\binom{3}{3}\binom{7}{0} = 1 \times 1 = 1$$

**Total de casos favorables:**  
$$63 + 21 + 1 = 85$$

**Probabilidad:**  
$$\frac{85}{120} = \boxed{\dfrac{17}{24}} \quad \text{(Mismo resultado)}$$
'''


```python
from math import comb
from IPython.display import display, Math
import random

# Cálculo teórico
P_teorico = 1 - comb(7, 3)/comb(10, 3)
display(Math(fr"P(\text{{Al menos 1 rojo}}) = 1 - \frac{{\binom{{7}}{{3}}}}{{\binom{{10}}{{3}}}} = \boxed{{\dfrac{{17}}{{24}}}} \approx {P_teorico*100:.2f}\%"))

# Simulación
def simular_extraccion(n_sim=100000):
    caja = ['N']*4 + ['A']*3 + ['R']*3
    exitos = 0
    for _ in range(n_sim):
        muestra = random.sample(caja, 3)
        if 'R' in muestra:
            exitos += 1
    return exitos/n_sim

P_sim = simular_extraccion()
display(Math(fr"\text{{Probabilidad simulada}} ({10^5}\text{{ iteraciones}}) = \boxed{{{P_sim*100:.2f}\%}}"))
```

# Probabilidad condicional de que un directivo sea comunicador
# =========================================================

## **Enunciado del Problema**
'''
En una empresa:
- $15\%$ son psicólogos (P)
- $10\%$ son comunicadores sociales (C)
- $25\%$ de los psicólogos son directivos
- $20\%$ de los comunicadores son directivos
- $10\%$ del resto (ni P ni C) son directivos

**Pregunta:**  
Si elegimos un empleado al azar y **es directivo**, ¿cuál es la probabilidad de que **sea comunicador**?
'''

## **Solución Teórica**
'''
### **Paso 1: Definir probabilidades base**
- $P(P) = 0.15$ (Psicólogos)
- $P(C) = 0.10$ (Comunicadores)
- $P(\text{Resto}) = 1 - 0.15 - 0.10 = 0.75$

### **Paso 2: Calcular probabilidades condicionales de ser directivo (D)**
- $P(D|P) = 0.25$ (Directivos entre psicólogos)
- $P(D|C) = 0.20$ (Directivos entre comunicadores)
- $P(D|\text{Resto}) = 0.10$ (Directivos en otros)

### **Paso 3: Aplicar Teorema de Bayes**
Queremos encontrar $P(C|D)$:

$$ P(C|D) = \frac{P(D|C) \cdot P(C)}{P(D)} $$

Donde $P(D)$ es la probabilidad total de ser directivo:

$$ P(D) = P(D|P)P(P) + P(D|C)P(C) + P(D|\text{Resto})P(\text{Resto}) $$

Sustituyendo valores:

$$ P(D) = (0.25 \times 0.15) + (0.20 \times 0.10) + (0.10 \times 0.75) $$
$$ P(D) = 0.0375 + 0.02 + 0.075 = 0.1325 $$

Ahora calculamos $P(C|D)$:

$$ P(C|D) = \frac{0.20 \times 0.10}{0.1325} = \frac{0.02}{0.1325} \approx 0.1509 $$

### **Respuesta Final:**
$$ P(\text{Comunicador}|\text{Directivo}) = \boxed{15.09\%} $$
'''

# Probabilidad condicional de que tome paracetamol dado que toma ibuprofeno
# =========================================================================

## **Enunciado del Problema**
'''
En la población de Turbaco:
- $48\%$ toma ibuprofeno (I)
- $33\%$ toma paracetamol (P)
- $22\%$ toma ambos medicamentos

**Pregunta:**  
Si una persona **toma ibuprofeno**, ¿cuál es la probabilidad de que **también tome paracetamol**?
'''

## **Solución Teórica**
'''
### **Definición de probabilidad condicional:**
$$ P(A \mid B) = \frac{P(A \cap B)}{P(B)} $$

### **Aplicación al problema:**
1. **Eventos:**
   - $P(I) = 0.48$ (Probabilidad de tomar ibuprofeno)
   - $P(P) = 0.33$ (Probabilidad de tomar paracetamol)
   - $P(I \cap P) = 0.22$ (Probabilidad de tomar ambos)

2. **Probabilidad condicional:**  
   $$ P(P \mid I) = \frac{P(I \cap P)}{P(I)} = \frac{0.22}{0.48} $$

3. **Resultado:**  
   $$ P(P \mid I) = \boxed{\dfrac{11}{24} \approx 45.83\%} $$
'''

# Solución al problema de las urnas con extracciones condicionadas
# ==============================================================

## **Enunciado del Problema**
'''
Tenemos 100 urnas de tres tipos:
1. **Tipo I:** 8 bolas blancas y 2 negras
2. **Tipo II:** 4 bolas blancas y 6 negras
3. **Tipo III:** 1 bola blanca y 9 negras

Se realizan dos extracciones con reemplazo:
1. Primera extracción: Bola blanca
2. Segunda extracción: Bola negra

Datos conocidos:
- $P(\text{Tipo I} \mid \text{Blanca}) = \dfrac{16}{39}$
- $P(\text{Tipo II} \mid \text{Negra}) = \dfrac{30}{61}$

**Objetivo:** Determinar el número de urnas del Tipo III ($N_3$).
'''

## **Solución Teórica**
'''
### **Paso 1: Definición de variables**
$$ N_1 + N_2 + N_3 = 100 $$

### **Paso 2: Aplicar Bayes para primera extracción (Blanca)**
$$ P(\text{Tipo I} \mid \text{Blanca}) = \frac{P(\text{Blanca} \mid \text{Tipo I}) \cdot P(\text{Tipo I})}{P(\text{Blanca})} $$

Donde:
$$ P(\text{Blanca}) = \frac{8}{10}N_1 + \frac{4}{10}N_2 + \frac{1}{10}N_3 $$

Resultando:
$$ \frac{16}{39} = \frac{\frac{8}{10}N_1}{\frac{8}{10}N_1 + \frac{4}{10}N_2 + \frac{1}{10}N_3} $$

Simplificando:
$$ 4N_2 + N_3 = 11.5N_1 \quad \text{(Ecuación 1)} $$

### **Paso 3: Aplicar Bayes para segunda extracción (Negra)**
$$ P(\text{Tipo II} \mid \text{Negra}) = \frac{P(\text{Negra} \mid \text{Tipo II}) \cdot P(\text{Tipo II})}{P(\text{Negra})} $$

Donde:
$$ P(\text{Negra}) = \frac{2}{10}N_1 + \frac{6}{10}N_2 + \frac{9}{10}N_3 $$

Resultando:
$$ \frac{30}{61} = \frac{\frac{6}{10}N_2}{\frac{2}{10}N_1 + \frac{6}{10}N_2 + \frac{9}{10}N_3} $$

Simplificando:
$$ 10N_1 + 45N_3 = 31N_2 \quad \text{(Ecuación 2)} $$

### **Paso 4: Resolver el sistema de ecuaciones**
1. De Ecuación 1: $N_3 = 11.5N_1 - 4N_2$
2. Sustituir en Ecuación 2:
   $$ 10N_1 + 45(11.5N_1 - 4N_2) = 31N_2 $$
   $$ 527.5N_1 = 211N_2 \Rightarrow N_2 = 2.5N_1 $$
3. Sustituir en $N_1 + N_2 + N_3 = 100$:
   $$ N_1 + 2.5N_1 + (11.5N_1 - 10N_1) = 100 $$
   $$ 5N_1 = 100 \Rightarrow N_1 = 20 $$
4. Hallar demás variables:
   $$ N_2 = 50, \quad N_3 = 30 $$

## **Respuesta Final**
El número de urnas del Tipo III es:
$$ \boxed{30} $$
'''

# Cálculo de probabilidad de que ninguno de los tres asesores aconseje comprar
# ============================================================================

## **Enunciado del Problema**
'''
Una empresa consulta a tres asesores financieros independientes sobre la compra de un paquete de acciones.
Sus probabilidades de recomendar la compra son:
- Asesor 1: $P(A_1) = 0.8$ (probabilidad de recomendar compra)
- Asesor 2: $P(A_2) = 0.5$
- Asesor 3: $P(A_3) = 0.3$

**Objetivo:**  
Calcular la probabilidad de que **ninguno** recomiende la compra.
'''

## **Solución Teórica**
'''
### **Paso 1: Calcular probabilidades de no recomendar**
Como son eventos independientes:
- Probabilidad de que Asesor 1 **no** recomiende:  
  $$ P(\overline{A_1}) = 1 - P(A_1) = 1 - 0.8 = 0.2 $$

- Probabilidad de que Asesor 2 **no** recomiende:  
  $$ P(\overline{A_2}) = 1 - 0.5 = 0.5 $$

- Probabilidad de que Asesor 3 **no** recomiende:  
  $$ P(\overline{A_3}) = 1 - 0.3 = 0.7 $$

### **Paso 2: Calcular probabilidad conjunta**
Como los asesores son independientes:
$$
P(\text{Ninguno recomiende}) = P(\overline{A_1}) \times P(\overline{A_2}) \times P(\overline{A_3}) = 0.2 \times 0.5 \times 0.7
$$

### **Resultado Final:**
$$
P(\text{Ninguno recomiende}) = \boxed{0.07} \text{ (7%)}
$$

# Probabilidad condicional de no estar cualificado dado que es universitario
# =========================================================================

## **Enunciado del Problema**
'''
Datos de una empresa de trabajo temporal:
- **Total solicitantes:**
  - $25\%$ cualificados
    - Distribución entre cualificados:
      - $20\%$ universitarios
      - $30\%$ Formación Profesional
      - $50\%$ Bachillerato
  - $75\%$ no cualificados
    - Distribución entre no cualificados:
      - $40\%$ universitarios
      - $40\%$ Formación Profesional
      - $20\%$ Bachillerato

**Pregunta:**  
¿Qué porcentaje de estudiantes **universitarios** que solicitaron empleo **no estaba cualificado**?
'''

## **Solución Teórica**
'''
### **Paso 1: Definir probabilidades base**
- $P(C) = 0.25$ (Probabilidad de estar cualificado)
- $P(\overline{C}) = 0.75$ (Probabilidad de no estar cualificado)

### **Paso 2: Calcular proporciones por nivel educativo**
1. **Entre cualificados ($C$):**
   - $P(U|C) = 0.20$ (Universitarios)
   - $P(FP|C) = 0.30$ (Formación Profesional)
   - $P(B|C) = 0.50$ (Bachillerato)

2. **Entre no cualificados ($\overline{C}$):**
   - $P(U|\overline{C}) = 0.40$ (Universitarios)
   - $P(FP|\overline{C}) = 0.40$ (Formación Profesional)
   - $P(B|\overline{C}) = 0.20$ (Bachillerato)

### **Paso 3: Aplicar Teorema de Bayes**
Queremos encontrar $P(\overline{C}|U)$:

$$ P(\overline{C}|U) = \frac{P(U|\overline{C}) \cdot P(\overline{C})}{P(U)} $$

Donde $P(U)$ es la probabilidad total de ser universitario:

$$ P(U) = P(U|C)P(C) + P(U|\overline{C})P(\overline{C}) $$
$$ P(U) = (0.20 \times 0.25) + (0.40 \times 0.75) = 0.05 + 0.30 = 0.35 $$

Ahora calculamos $P(\overline{C}|U)$:

$$ P(\overline{C}|U) = \frac{0.40 \times 0.75}{0.35} = \frac{0.30}{0.35} \approx 0.8571 $$

### **Respuesta Final:**
$$ P(\text{No cualificado} \mid \text{Universitario}) = \boxed{85.71\%} $$
'''
